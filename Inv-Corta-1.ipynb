# Example of data filtering and processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import preprocessing

# creation of dataset
header = ['age','bp','sg','al','su','rbc','pc','pcc',
    'ba','bgr','bu','sc','sod','pot','hemo','pcv',
    'wbcc','rbcc','htn','dm','cad','appet','pe','ane',
    'classification']

ds = pd.read_csv("data/chronic_kidney_disease.arff",
                        header=None,
                        names=header)

# replace ? into Nan
ds = ds.replace('?', np.nan)
# drop the NaN
ds = ds.dropna(axis=0, how="any")

targets = ds['classification'].astype('category')

# red: disease,  green: no disease
label_color = ['red' if i == 'ckd' else 'green' for i in targets]

# list of categorical features to drop
categories = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'classification']

# drop the "categorical" features
ds.drop(categories, axis=1, inplace=True)

# StandardScaler: mean=0, variance=1
ds = preprocessing.StandardScaler().fit_transform(ds)

pca = PCA(n_components=2)
pca.fit(ds)

# transformed data
T = pca.transform(ds)
# change 'T' to Pandas-DataFrame to plot using Pandas-plots
T = pd.DataFrame(T)

# plot the data
T.columns = ['PCA component 1', 'PCA component 2']
T.plot.scatter(x='PCA component 1', y='PCA component 2', marker='o',
               alpha=0.7, # opacity
               color=label_color,
               title="red: ckd, green: not-ckd" )
plt.show()
```
